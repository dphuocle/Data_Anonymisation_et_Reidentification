```python
grouped_data_by_id = df.groupby("Id")
```

```python
for name, group in grouped_data_by_id:
    plt.scatter(group['Longitude'], group['Latitude'])
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.title(f'Heatmap by Id ={name}')
    plt.show()
```
