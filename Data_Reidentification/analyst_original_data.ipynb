```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('real_data_set.csv', sep = "\t", names = ['Id', 'DateAndTime', 'Longitude', 'Latitude'])
```

```python
df['DateAndTime'] = pd.to_datetime(df['DateAndTime'], errors = 'coerce')
```

```python
df['Week'] = df['DateAndTime'].dt.isocalendar().week
```

```python
df
```

```python
week_10_data = df[df['Week'] == 10]
```

```python
week_10_data
```

```python
week_10_data_unique_id = week_10_data['Id'].sort_values(ascending = True).unique()
```

```python
print(week_10_data_unique_id)
print('Number of Unique Id of Week 10: ', week_10_data['Id'].nunique())
```

```python
week_11_data = df[df['Week'] == 11]
week_11_data
```

```python
week_11_data_unique_id = week_11_data['Id'].sort_values(ascending = True).unique()
```

```python
print(week_11_data_unique_id)
print('Number of Unique Id of Week 11: ', week_11_data['Id'].nunique())   
```
