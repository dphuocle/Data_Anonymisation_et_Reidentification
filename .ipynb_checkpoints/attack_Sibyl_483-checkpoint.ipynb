```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('real_data_set.csv', sep = "\t", names = ['Id', 'Date and Time', 'Longitude', 'Latitude'])
```

```python
grouped = df.groupby("Id")
for name, group in grouped:
    plt.plot(group['Longitude'], group['Latitude'])
    plt.legend()
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.title(f'Plot by id = {name}')
    plt.show()
```
